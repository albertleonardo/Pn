Demo data for the global database of regional seismograms

```python
import h5py
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image

plt.rcParams['font.family']='Helvetica'
plt.rcParams['font.size']=18
```

The following dataset contains wavefroms for which both a P? phase and an S? phase were found in the catalog

```python
f = h5py.File('demodata.hdf5', 'r')
demodata = f['data']
```

 Each subgroup contains information regarding one trace. It contains the attributes shown below

```python
one_demo = demodata[list(demodata.keys())[12]]
for key in list(one_demo.attrs.keys()):print(key)
```

 The waveforms are 5 minutes long, bandpassed filtered 1-45 Hz, with a random shift, the first arriving phase will be sometime between second 10 and 60. Waveforms are also resampled at 100 Hz,
